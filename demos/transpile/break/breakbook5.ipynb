```python
import pyzx
from pyzx.circuit.qasmparser import QASMParser
```

```python
qasm = """
OPENQASM 2.0;
include "qelib1.inc";
qreg qr[2];
h qr[0];
h qr[1];
h qr[0];
s qr[0];
h qr[1];
s qr[1];
cx qr[0],qr[1];
cx qr[1],qr[0];
h qr[0];
s qr[0];
h qr[1];
s qr[1];
x qr[0];
x qr[1];
"""
print(qasm)
```

```python
p = QASMParser()
circ = p.parse(qasm)
print(circ.__dict__)
```

```python
g = circ.to_graph()
pyzx.draw(g)
```

```python
pyzx.full_reduce(g)
pyzx.draw(g)
```

```python
new_circ = pyzx.extract.streaming_extract(g)
new_circ = pyzx.optimize.basic_optimization(new_circ.to_basic_gates())
print(new_circ.__dict__)
```

```python
assert(pyzx.compare_tensors(circ, new_circ, False))
```
